# 4. Data Cleaning

## Fixing the data types

```python
# Converting TotalCharges and MonthlyCharges to numeric data types

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
```

```python
replace_cols = [
    'MultipleLines','OnlineSecurity','OnlineBackup',
    'DeviceProtection','TechSupport','StreamingTV','StreamingMovies'
]

for col in replace_cols:
    df[col] = df[col].replace({
        'No internet service':'No',
        'No phone service':'No'
    })
```

# 5. FUTURE ENGINEEERING

```python
# 1. Flag for customers lacking key services (for Top 5 Churn Reasons chart)

df['NoTechSupport'] = (df['TechSupport'] == 'No').astype(int) 

df['NoOnlineSecurity'] = (df['OnlineSecurity'] == 'No').astype(int) 

df['MonthToMonth'] = (df['Contract'] == 'Month-to-month').astype(int) 

# 2. High monthly charge flag (over $80)
df['HighMonthlyCharge'] = (df['MonthlyCharges'] > 80).astype(int)

# 3. Electronic check flag
df['ElectronicCheck'] = (df['PaymentMethod'] == 'Electronic check').astype(int)
```

```python
# Mapping the values 1s and 0s in the SeniorCitizen column to Yes and No respectively.
Mapping = {1:"Yes", 0:"No"}

df["SeniorCitizen"] = df["SeniorCitizen"].map(Mapping)
```

```python
# Mapping churn
df['Churn_Flag'] = df['Churn'].map({'Yes': 1, 'No': 0}) 

# 5. Churn flag (0/1 for calculations)
df['Churn_Flag'] = (df['Churn'] == 'Yes').astype(int)
```

```python
# Creating a column for high risk customers
df['HighRiskSegment'] = (
    (df['Contract'] == 'Month-to-month') & 
    (df['TechSupport'] == 'No') & 
    (df['MonthlyCharges'] > 70)
).astype(int)

Mapping = {1:"Yes", 0:"No"}

df["HighRiskSegment"] = df["HighRiskSegment"].map(Mapping)
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0,12,24,48,60,72],
    labels=['0-12','12-24','24-48','48-60','60-72']
)
```

```python
# Monthly charge segmentation
df['ChargeLevel'] = pd.cut(
    df['MonthlyCharges'],
    bins=3,
    labels=['Low','Medium','High']
)
```

```python
# Total Services available to customers\
services = [
    'PhoneService','OnlineSecurity','OnlineBackup',
    'DeviceProtection','TechSupport','StreamingTV','StreamingMovies'
]

df['TotalServices'] = (df[services] == 'Yes').sum(axis=1)
```
